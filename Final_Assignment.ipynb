# Data loading:

```python
with open('../assessments/config.yaml', 'r') as stream:
    config= yaml.safe_load(stream)
total_death= config['Total_death']
df_total= pd.read_csv(total_death, sep=';')
df_total.head()
```

```python
death_cause = config['couse_of_death']
df_cause = pd.read_csv(death_cause, sep= ';')

df_cause.head()

```

```python
#drop the non infectiouse deases:
df_cause.drop(df_cause.iloc[:,13:16], axis=1, inplace=True)
df_cause.head()

```

```python
df_cause.drop(df_cause.iloc[:,9:11], axis=1, inplace=True)
df_cause.head()
```

```python
#changing the name of the columns to make it more readble
# here i make a dictinary of old name as a key and new  as value:
new_columns ={'Sex':'Gender','k_11Tuberculosis_3':'Tuberculosis','k_12CerebrospinalInfection_4':'CerebrospinalInfection',
             'k_13ViralHepatitis_5':'ViralHepatitis','k_14AIDS_6':'AIDS', 
              'k_15OtherInfectAndParasiticDiseases_7':'OtherInfectAndParasiticDiseases','k_81Influenza_51':'Influenza','k_82Pneumonia_52':'Pneumonia', 
             'k_181ConfirmedCOVID19_94':'ConfirmedCOVID19'}
df_cause.rename(columns= new_columns, inplace=True)

df_cause.head()




```

```python
#Periods value is object type, i should change it to datetime formate, i have to replace the unknown-string-format-on-pd-to_datetime:      

df_cause['Periods']=df_cause['Periods'].str.replace('JJ00','')

df_cause['Periods']= pd.to_datetime(df_cause['Periods'])
df_cause.dtypes
```

```python
# changeing the object data type to datetime formate
df_total['Periods']= df_total['Periods'].str.replace('JJ00','')
df_total['Periods']=pd.to_datetime(df_total['Periods'])
df_total.dtypes
```

```python
# #collecting data from 2015
new_df_total = df_total[df_total['Periods']>'2014']
dict_total_death= {'Age31December':'Age','Deaths_1':'Total_death'}
#cahinging column name
new_df_total.rename(columns=dict_total_death, inplace= True)
new_df_total.drop(columns=['ID'],inplace = True)
new_df_total.head()
```

```python
new_df_cause = df_cause[df_cause['Periods']>'2014']
new_df_cause.drop(columns=['ID'],inplace = True)
new_df_cause.head()
```
